```
# mengimport library yang dibutuhkan
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```
# Membuat objek untuk menampung datasets
mnist = tf.keras.datasets.mnist
# Membagi datasets menjadi data training dan testing
(train_pic, train_label) , (testing_pic, testing_label) = mnist.load_data()
```

```
np.set_printoptions(linewidth=200)
plt.imshow(train_pic[1])
print(train_label[1])
```

```
# Menormalisasi gambar dengan membagi setiap piksel sebesar 255
# Karena sebuah pixel berkisar dari 0 sampai 255
train_pic = train_pic/255.0
train_label = train_label/255
```

```
# Membuat model neural network dengan menggunakan layer Flatten diawal
# Untuk mengubah matriks 2 dimensi menjadi 1 dimensi
model = Sequential([
        Flatten(input_shape = (28,28)),
        Dense(units = 128, activation = 'relu'),
        Dense(units = 10, activation = 'softmax')
])
```

```
# Menentukan optimizer dan loss untuk model
model.compile(
    optimizer = 'Adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

# Melatih mesin mengenali datasets
model.fit(
    train_pic,
    train_label,
    epochs = 50
)
```
