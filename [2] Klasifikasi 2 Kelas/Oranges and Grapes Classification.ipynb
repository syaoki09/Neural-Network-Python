```
# Mengimport library yang diperlukan
import pandas as pd
from google.colab import files
import io
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```
# Mengupload file 
uploaded = files.upload()
```

```
# Mengimport datasets dan mengubah ke dalam data frame
df = pd.read_csv(io.BytesIO(uploaded['citrus.csv']))
df.head()
```

```
# karena Neural Netwoek tidak bisa memproses String
# Maka akan diubah terlebih dahulu ke bentuk numerik
df.name[df.name == 'orange'] = 0.0
df.name[df.name == 'grapefruit'] = 1.0
```

```
# Karena datasets masih dalam bentuk dataframe
# Maka datasets harus ke dalam bentuk array agar bisa di proses
datasets = df.values
```

```
# Memisahkan datasets antara atribut dan label/kelas
atribut = datasets[:,1:6]
label = datasets[:,0]
```

```
# Menormalisasi data pada atribut 
# agar model dapat mempelajari datasets dengan baik
min_max_scaler = preprocessing.MinMaxScaler()
atribut = min_max_scaler.fit_transform(atribut)
```

```
# Memisah data menjadi data train dan testing sebanyak 30%
x_train, x_test, y_train, y_test = train_test_split(atribut, label, test_size = 0.3)
```

```
# Mengubah tipe data integer menjadi float agar dapat di proses di Neural Network
y_train = y_train.astype(np.float)
y_test = y_test.astype(np.float)
```

```
# Membuat model neural network dengan 3 layers dengan output antara 0-1
model = Sequential([
        Dense(units = 32, activation='relu', input_shape = [5,]),
        Dense(units = 32, activation='relu'),
        Dense(units = 1, activation = 'sigmoid')  
])
# Catatan : Sigmoid cocok digunakan pada masalah klasifikasi biner (0-1)
```

```
# Memberikan optimizer, loss function untuk model belajar dengan baik
model.compile(
    optimizer = 'sgd',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
# Catatan : 
# sgd (Stochastic gradient descent) cocok dipakai pada datasets berukuran kecil
# binary_crossentropy sesuai dengan kasus 2 kelas/label (biner)
# accuracy pada metrics berguna menampikkan akurasi pada setiap proses pelatihan
```

```
# Melatih model dengan datasets

model.fit(
    x_train,
    y_train,
    epochs = 150
)
```

```
# Mengevaluasi model dengan melihat apakah model underfit atau overfit
model.evaluate(x_test,y_test)
```

fungsi evaluate mengembalikan 2 nilai, yang pertama adalah nilai loss dan yang kedua nilai akurasi
