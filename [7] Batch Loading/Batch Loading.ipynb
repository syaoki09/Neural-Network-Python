```
# Mengimport library yang dibutuhkan
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Flatten, Dense
```

```
# Membuat objek datasets
mnist = tf.keras.datasets.mnist
```

```
# Membagi datasets menjadi data training dan data testing
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
# Normalisasi gambar dengan membagi dengan 255
x_train = x_train/255
y_train = y_train/255
```

```
# Membuat model neural network
model = Sequential([
        Flatten(input_shape = (28,28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
])
```

```
# Memberikan optimizer dan loss supaya model dapat belajar dengan baik
model.compile(
    optimizer = 'Adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```
# Pembelajaran mesin dari datasets tanpa menggunakan batch size
model.fit(
    x_test,
    y_test,
    epochs = 5
)
```

```
# Pembelajaran mesin dari datasets dengan menggunakan batch size
model.fit(
    x_test,
    y_test,
    batch_size = 64,
    epochs = 5
)
```
