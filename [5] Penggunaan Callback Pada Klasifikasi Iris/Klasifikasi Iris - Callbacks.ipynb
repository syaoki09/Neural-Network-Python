```
# mengimport library yang dibutuhkan
import tensorflow as tf
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```
from google.colab import files
import io
uploaded = files.upload()
```

```
# Mengimport datasets menjadi dataframe
df = pd.read_csv(io.BytesIO(uploaded['Iris.csv']))
df.head()
```

```
# Membuang kolom id karena tidak ada hubungannya ke label
df = df.drop(columns = 'Id')
```

```
# Melakukan One Hot Encoding karena label berupa kategorikal
category = pd.get_dummies(df.Species)
category
```

```
# Menggabungkan dataframe dengan hasil dari encoding
new_df = pd.concat([df, category], axis = 1)
# Membuang kolom species yang berupa kategorikal
new_df = new_df.drop(columns = 'Species')
new_df
```

```
# Mengubah dataframe menjadi array
datasets = new_df.values
datasets
```

```
# Memisah atribut dan label pada datasets
atribut = datasets[:,0:4]
label = datasets[:,4:7]
```

```
# Normalisasi pada atribut agar model bisa belajar dengan baik
min_max_scaler = preprocessing.MinMaxScaler()
x_scale = min_max_scaler.fit_transform(atribut)
x_scale
```

```
# Memisah data training dan data test
x_train, x_test, y_train, y_test = train_test_split(
    x_scale,
    label,
    test_size = 0.3
)
```

```
# Membuat model neural network
model = Sequential([
          Dense(units = 64, activation='relu',input_shape = (4,)),
          Dense(units = 64, activation='relu'),
          Dense(units = 3, activation = 'sigmoid')
])
```

```
model.compile(
    optimizer = 'Adam',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)
```

Membuat kelas callback untuk memberitahukan dan memberhentikan model untuk belajar ketika suatu kondisi terpenuhi

```
# Membuat callback untuk memberhentikan jika akurasi mencapai 90%
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy') > 0.9):
      print('Akurasi telah mencapai >90%!')
      self.model.stop_training = True

callbacks = myCallback()
```

```
# Pembelajaran mesin menggunakan callback
model.fit(
    x_train,
    y_train,
    epochs = 50,
    callbacks = [callbacks]
)
```
